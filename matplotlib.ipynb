```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#line plot
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('x vs y')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
```

```python
#Adding two line in the same plot
x
```

```python
y1=2*x
```

```python
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=3)
plt.plot(x,y2,color='r',linewidth=2)
plt.title("Line plot")
plt.xlabel("x-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
#adding sub plot
x
```

```python
y1
```

```python
y2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y2,color='y',linewidth=3)
plt.grid(True)
plt.show()
```

```python
#Bar plot
dict1= {"Asf":80,"Max":56,"Sam":27}
```

```python
names = list(dict1.keys())
```

```python
values = list(dict1.values())
```

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
```

```python
#horizontal bar plot
plt.barh(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
```

```python
#scatter plot
x=[10,20,30,40,50,60,70,80,90]
y=[8,1,7,2,0,3,7,3,2]
```

```python
plt.scatter(x,y,color='r',marker="*",s=100)
plt.title("x vs y")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
```

```python
#Adding two markers in the same scatter plot

x
```

```python
y
```

```python
a=[9,8,0,5,6,3,1,8,0]
```

```python
a
```

```python
plt.scatter(x,y,color='g',marker='*',s=100)
plt.scatter(x,a,color='y',marker='.',s=200)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
```

```python
#Adding subplot
plt.subplot(1,2,1)
plt.scatter(x,y,color='g',marker='*',s=100)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)

plt.subplot(1,2,2)
plt.scatter(x,a,color='y',marker='.',s=200)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
```

```python
#Histogram
data= [1,3,3,3,3,9,9,5,4,4,8,8,8,6,7]
```

```python
plt.hist(data,color='g',bins=5)
plt.show()
```

```python
#Box-Plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]

data1=list([one,two,three])
plt.boxplot(data1)
plt.show()
```

```python
#violin plot
data1
```

```python
plt.violinplot(data1,showmedians=True)
plt.show()
```

```python
#pie-chart
fruit=['Apple','Orange','Mango','Guava']
quantity=[67,34,100,29]

plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','grey','blue','green'])
plt.show()
```

```python
#DoughNut-Chart
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.1f%%')
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
